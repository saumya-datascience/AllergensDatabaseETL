```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
from allergy_links import soybean_links
from functions import get_products, flatten_list
```

```python
all_soybean = [get_products(i) for i in soybean_links]
all_soybean
```

```python
all_soybean_list = flatten_list(all_soybean)
all_soybean_list
```

```python
soybean_df = pd.DataFrame(all_soybean_list)
soybean_df.rename(columns={0: "products"}, inplace=True)
ingredient = "soybean"
soybean_df["ingredient"] = ingredient
soybean_df = soybean_df[["ingredient", "products"]]
soybean_df
```

```python
soybean_df["products"] = soybean_df["products"].map(str.title)
soybean_df
```

```python
soybean_df_1 = soybean_df.drop_duplicates()
soybean_df_1.reset_index(inplace=True)
```

```python
soybean_df_1.head()
```

```python
soybean_df_1.drop("index", inplace=True, axis=1)
```
