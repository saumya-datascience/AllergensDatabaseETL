```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = "https://acaai.org/allergies/types/food-allergy"
url_path = requests.get(url)
```

```python
soup = bs(url_path.content)
# soup
```

```python
symptoms_link = soup.find_all("div", class_="field-items")
# symptoms_link
```

```python
hrefs_links = []
for i in symptoms_link:
    for a in i.find_all('a', href=True): 
        if a.text:
            hrefs_links.append(a['href'])
hrefs_links[3:11]
```

```python
allergens = ["egg", "milk", "peanuts", "tree nuts", "fish", "shellfish", "wheat", "soy"]
```

```python
full_links = ["https://acaai.org" + i for i in hrefs_links[3:11]]
full_links
```

```python
def symptoms(url):
    url_path = requests.get(url)
    soup = bs(url_path.content)
    soup = bs(url_path.content)
    symptoms_1 = soup.find_all("div", class_="field-item even")
    symptoms_list = [i.text for i in symptoms_1[4].find_all("li")]
    return symptoms_list
```

```python
all_symp = [symptoms(i) for i in full_links]
all_symp
```

```python
result = {allergens[i]: all_symp[i] for i in range(len(allergens))}
result
```

```python
symptoms_df = pd.DataFrame.from_dict(result, orient='index')
symptoms_df
```
