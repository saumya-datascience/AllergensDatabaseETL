```python
import pandas as pd
```

```python
file_path_1 = "../Allergens-in-Goya-Food-Products-/data/fish_substitute.csv"
file_path_2 = "../Allergens-in-Goya-Food-Products-/data/egg_subs.csv"
file_path_3 = "../Allergens-in-Goya-Food-Products-/data/shellfish_substitutes.csv"
file_path_4 = "../Allergens-in-Goya-Food-Products-/data/soybean_substitute.csv"
file_path_5 = "../Allergens-in-Goya-Food-Products-/data/milk_sub.csv"
file_path_6 = "../Allergens-in-Goya-Food-Products-/data/wheat_sub.csv"
file_path_7 = "../Allergens-in-Goya-Food-Products-/data/peanut_sub.csv"
file_path_8 = "../Allergens-in-Goya-Food-Products-/data/treenut_sub.csv"
```

```python
fish_sub = pd.read_csv(file_path_1)
```

```python
egg_sub = pd.read_csv(file_path_2)
```

```python
shellfish_sub = pd.read_csv(file_path_3)
```

```python
soybean_sub = pd.read_csv(file_path_4)
```

```python
milk_sub = pd.read_csv(file_path_5)
```

```python
wheat_sub = pd.read_csv(file_path_6)
```

```python
peanut_sub = pd.read_csv(file_path_7)
```

```python
treenut_sub = pd.read_csv(file_path_8)
```

```python
all_substitutes = pd.concat([fish_sub, egg_sub, shellfish_sub, 
                          soybean_sub, milk_sub, wheat_sub, peanut_sub, treenut_sub])
```

```python
all_substitutes.drop("Unnamed: 0", inplace=True, axis=1)
```

```python
all_substitutes
```

```python
file_path_9 = "../Allergens-in-Goya-Food-Products-/data/allergens.csv"
allergens_table = pd.read_csv(file_path_9)
allergens_table
```

```python
allergens_table.rename(columns={"allergens": "ingredient"}, inplace=True)
```

```python
allergens_table
```

```python
all_substitutes_df = pd.merge(all_substitutes, allergens_table, on="ingredient")
all_substitutes_df
```

```python
all_substitutes_df = all_substitutes_df[["allergy_id", "ingredient", "substitute"]]
all_substitutes_df
```

```python
all_substitutes_df.to_csv("data/all_substitutes.csv")
```
