```python
import pandas as pd
```

```python
file_path_1 = "../Allergens-in-Goya-Food-Products-/data/fish.csv"
file_path_2 = "../Allergens-in-Goya-Food-Products-/data/egg.csv"
file_path_3 = "../Allergens-in-Goya-Food-Products-/data/shellfish.csv"
file_path_4 = "../Allergens-in-Goya-Food-Products-/data/soybean.csv"
file_path_5 = "../Allergens-in-Goya-Food-Products-/data/milk-final.csv"
file_path_6 = "../Allergens-in-Goya-Food-Products-/data/wheat-final.csv"
file_path_7 = "../Allergens-in-Goya-Food-Products-/data/peanut.csv"
file_path_8 = "../Allergens-in-Goya-Food-Products-/data/treenut.csv"
```

```python
fish_table = pd.read_csv(file_path_1)
```

```python
egg_table = pd.read_csv(file_path_2)
```

```python
shellfish_table = pd.read_csv(file_path_3)
```

```python
soybean_table = pd.read_csv(file_path_4)
```

```python
milk_table = pd.read_csv(file_path_5)
```

```python
wheat_table = pd.read_csv(file_path_6)
```

```python
peanut_table = pd.read_csv(file_path_7)
```

```python
treenut_table = pd.read_csv(file_path_8)
```

```python
all_products = pd.concat([fish_table, egg_table, shellfish_table, 
                          soybean_table, milk_table, wheat_table, peanut_table, treenut_table])
```

```python
all_products
```

```python
all_products.reset_index(inplace=True)
```

```python
all_products.head()
```

```python
all_products.drop(["index", "Unnamed: 0"], inplace=True, axis=1)
```

```python
all_products
```

```python
file_path_9 = "../Allergens-in-Goya-Food-Products-/data/allergens.csv"
```

```python
allergens_table = pd.read_csv(file_path_9)
allergens_table
```

```python
allergens_table.rename(columns={"allergens": "ingredient"}, inplace=True)
```

```python
allergens_table
```

```python
all_products_df = pd.merge(all_products, allergens_table, on="ingredient")
all_products_df
```

```python
all_products_df = all_products_df[["allergy_id", "ingredient", "products"]]
all_products_df
```

```python
all_products_df.to_csv("data/all_products.csv")
```
