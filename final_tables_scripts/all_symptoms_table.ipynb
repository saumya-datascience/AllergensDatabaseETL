```python
import pandas as pd
```

```python
file_path = "../Allergens-in-Goya-Food-Products-/data/allergens_symptoms.csv"
```

```python
symptoms_table = pd.read_csv(file_path)
symptoms_table
```

```python
symptoms_table.drop("Unnamed: 0", inplace=True, axis=1)
```

```python
symptoms_table.rename(columns={"allergens": "ingredient"}, inplace=True)
symptoms_table
```

```python
symptoms_table["ingredient"].replace({"tree nuts": "treenut", "peanuts": "peanut", "soy": "soybean"}, inplace=True)
symptoms_table
```

```python
file_path_2 = "../Allergens-in-Goya-Food-Products-/data/allergens.csv"
allergens_table = pd.read_csv(file_path_2)
allergens_table
```

```python
allergens_table.rename(columns={"allergens": "ingredient"}, inplace=True)
allergens_table
```

```python
all_symptoms_df = pd.merge(symptoms_table, allergens_table, on="ingredient")
all_symptoms_df
```

```python
all_symptoms_df = all_symptoms_df[["allergy_id", "ingredient", "symptoms"]]
all_symptoms_df
```

```python
all_symptoms_df.to_csv("data/all_symptoms.csv")
```

```python
all_symptoms_df1 = all_symptoms_df.drop("ingredient", axis=1)
all_symptoms_df1
```

```python
all_symptoms_df1.to_csv("data/all_symptoms_new.csv")
```
